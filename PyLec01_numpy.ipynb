# numpy

## How to import 'numpy'

```python
import numpy as np
```

Numpy is a python libray for mathematical computation.

We can create an array using numpy.

```python
x = np.array([1, 2, 3])
print(x)
```

Numpy has elementary functions such as exp, sin, cos.

To draw a graph we need **matplotlib** library.

For example, let us draw $y=sin(x)$ for $x\in[-2\pi,2\pi]$.

```python
x = np.linspace(-2*np.pi,2*np.pi)
y = np.sin(x)

import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

For example, we can draw $y=e^x$ for $x\in[0,10]$.

```python
x = np.arange(0,10,.1)
y = np.exp(x)
plt.plot(x,y)
plt.show()
```
